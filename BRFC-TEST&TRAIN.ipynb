```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
from pathlib import Path

from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python
# Load the data
file_path = Path('C:/Users/jyots/Desktop/final/train_dataset.csv')
train_df = pd.read_csv(file_path)

# Drop the null columns where all values are null
train_df = train_df.dropna(axis='columns', how='all')

# Drop the null rows
train_df = train_df.dropna()


train_df.head()
```

```python
# Load the data
file_path = Path('C:/Users/jyots/Desktop/final/test_dataset.csv')
test_df = pd.read_csv(file_path)

test_df.head()
```

```python
df = pd.concat([train_df, test_df], axis=0, join="outer")
df

non_smokers = df[df.smoking==0]
smokers = df[df.smoking==1]
new_df = pd.concat([non_smokers,smokers])
new_df
```

```python
new_df.dtypes
```

```python
X = new_df[new_df.smoking != "smoking"]
y = pd.DataFrame(new_df["smoking"])
```

```python
X.describe()
```

```python
y['smoking'].value_counts()
```

```python
from collections import Counter
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1)
X_train.shape
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.datasets import make_classification
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
brf = BalancedRandomForestClassifier(n_estimators=100, random_state=0)
brf
```

```python
brf.fit(X_train, y_train)
y_pred = brf.predict(X_test)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score
y_pred = brf.predict(X_test)
y_pred
```

```python
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
y_pred = brf.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```
